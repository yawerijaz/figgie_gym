```python
import random

from figgie_gym.market.common import Quantity, Symbol, Price, Side
from figgie_gym.market.market import Market
from figgie_gym.market.utilities import MarketDisplay


from figgie_gym.market.common import OrderType
symbol1 = Symbol('10001')
symbol2 = Symbol('20002')


class Simulator:
    def __init__(self):
        self.market = Market([symbol1, symbol2])
        self.client1 = self.market.new_client('client1',{symbol1: Quantity(20), symbol2: Quantity(20)})
        self.client2 = self.market.new_client('client2',{symbol1: Quantity(20), symbol2: Quantity(20)})
        self.tape = self.market.create_trade_ticker_tape()

        self.manual_client = self.market.new_client('manual',{symbol1: Quantity(9999), symbol2: Quantity(20)})
        self.display = MarketDisplay(self.market, self.tape, "market_snapshot.txt")

    def manual_trade(self, side: Side, symbol: Symbol, price: Price, quantity: int = 1, order_type:OrderType=OrderType.IMMEDIATE_OR_CANCEL):
        self.manual_client.send_order(symbol, side, price, Quantity(quantity), order_type=order_type)
        self.display.print()

    def tick(self):
        for _ in range(10000):
            try:
                client = random.choice([self.client1, self.client2])
                symbol = random.choice([symbol1])
                side = random.choice([Side.BUY, Side.SELL])
                price = Price(random.randrange(1, 20))
                client.send_order(symbol, side, price)
                print(client, symbol, side.name, price)
                self.display.print()
                yield
            except Exception as ex:
                print(type(ex), ex)

sim = Simulator()
```
